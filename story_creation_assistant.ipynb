```python
from langchain_core.prompts import PromptTemplate
from langchain_upstage import ChatUpstage
from langchain_core.output_parsers import StrOutputParser

llm = ChatUpstage()

prompt_file_path = "prompt.txt"

with open(prompt_file_path, "r", encoding="UTF8") as file:
    content = file.read()
prompt_template = PromptTemplate.from_template(
    content + "{context}"
)

additional_prompt_file_path = "prompt2.txt"
with open(additional_prompt_file_path,"r",encoding="UTF8") as file:
    additional_prompt_content = file.read()
additional_prompt =additional_prompt_content + "{additional_context}"
full_prompt_template = PromptTemplate.from_template(
    prompt_template.template + additional_prompt
)
chain = full_prompt_template | llm | StrOutputParser()
context = """
너만의 아포칼립스 세계관을 만들어줘
"""
additional_context = """
추가적으로 너가 만든 세계관을 바탕으로 너의 세계관의 장소를 자세하게 만들어줘
"""
output = chain.invoke({"context": context,"additional_context":additional_context})

    
with open('worldview.txt', 'w',encoding="utf8") as f:
    f.write(output)

with open('worldview.txt', 'r', encoding="utf8") as f:
    print(f.read())
# print("------------------")
# print(chain.invoke({"context": output}))
```

```python
from langchain_core.prompts import PromptTemplate
from langchain_upstage import ChatUpstage
from langchain_core.output_parsers import StrOutputParser

llm = ChatUpstage()

file_path = "character_prompt.txt"

with open(file_path, "r", encoding="UTF8") as file:
    content = file.read()
prompt_template = PromptTemplate.from_template(
    content + "{context}"
)

with open("worldview.txt","r",encoding="UTF8") as file:
    worldview = file.read()

chain = prompt_template | llm | StrOutputParser()
context = "세계관은 다음과 같다.\n"+ worldview +" \n 세계관을 한번 더 생각해보자. \n세계관에 알맞는 인물을 정보를 꼭 내 프롬포트에 맞춰서 출력해줘"
output = chain.invoke({"context": context})
print(output)
with open('character.txt', 'w',encoding="utf8") as f:
    f.write(output)

with open('character.txt', 'r',encoding="utf8") as f:
    print(f.read())
print("------------------")
# print(chain.invoke({"context": output}))
```
